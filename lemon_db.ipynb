```python
# Using this to hide username and password

import os 
from dotenv import load_dotenv

load_dotenv()

username = os.getenv('USERNAME')
password = os.getenv('PASSWORD')
print(username)
print(password)
```

```python
#import mysql connector to establish a connection with mysql 

import mysql.connector as connector

try:
    connection = connector.connect(user=username,password=password)
    print('connection successful!')
except:
    print('connection not established')
```

```python
# Now that we have created a connection we need make cursor to communicate with mysql

cursor = connection.cursor()

# Create database and set it and use it 
cursor.execute("CREATE DATABASE little_lemon_db")
cursor.execute("USE little_lemon_db")

print("DB HAS BEEN MADE")
```

```python
# Need tables to insert data 

create_menuitem_table = """
CREATE TABLE MenuItems(
ItemID INT AUTO_INCREMENT,
Name VARCHAR(200),
Type VARCHAR(100),
Price INT,
PRIMARY KEY (ItemID)
);
"""

create_menu_table = """
CREATE TABLE MENUS(
MenuID INT,
ItemID INT,
Cuisine VARCHAR(100),
PRIMARY KEY (MenuID,ItemID)
);
"""

create_booking_table = """
CREATE TABLE Bookings(
BookingID INT AUTO_INCREMENT,
TableNo INT,
GuestFirstName VARCHAR(100) NOT NULL,
GuestLastName VARCHAR(100) NOT NULL,
BookingSlot TIME NOT NULL,
EmployeeID INT,
PRIMARY KEY (BookingID)
);
"""

create_orders_table = """
CREATE TABLE Orders(
OrderID INT,
TableNo INT,
MenuID INT,
BookingID INT,
BillAmount INT,
Quantity INT,
PRIMARY KEY (OrderID,TableNo)
);
"""

create_employees_table = """
CREATE TABLE Employees(
EmployeeID INT AUTO_INCREMENT PRIMARY KEY,
Name VARCHAR(255),
Role VARCHAR(100),
Address VARCHAR(255),
Contact_Number INT,
Email VARCHAR(255),
Annual_Salary VARCHAR(100)
);
"""
```

```python
# Create the tables 
cursor.execute(create_menuitem_table)
cursor.execute(create_menu_table)
cursor.execute(create_booking_table)
cursor.execute(create_orders_table)
cursor.execute(create_employees_table)

print('Table been created')
```
